```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import re
```

```python
cat = ["laptops","smart+phones","headphones"]
title=[]
sec=[]
loc=[]
price=[]
shipping=[]
rating=[]
review=[]
bids=[]
category=[]

for j in cat:
    for i in range(1,21):
        URL = 'https://www.ebay.com/sch/i.html?_from=R40&_nkw='+j+'&_sacat=0&_pgn='+str(i)
        page = requests.get(URL).text
        soup = BeautifulSoup(page, 'lxml')

        for items in soup.find_all('li', class_='s-item'):
            try:  
                item_title = items.find('h3', class_='s-item__title').text
            except Exception as e:
                item_title = 'None'
            title.append(item_title)

            try:  
                item_sec = items.find('span', class_='SECONDARY_INFO').text
            except Exception as e:
                item_sec = 'None'
            sec.append(item_sec)

            try:  
                item_origin = items.find('span', class_='s-item__location s-item__itemLocation').text
                item_origin = re.sub('From ', '', item_origin)
            except Exception as e:
                item_origin = 'None'
            loc.append(item_origin)

            try:
                item_price = items.find('span', class_='s-item__price').text.split(" ")[1]
            except Exception as e:
                item_price = 'None'
            price.append(item_price)

            try:
                item_shipping = items.find('span', class_='s-item__shipping s-item__logisticsCost').text.split(" ")[1]
            except Exception as e:
                item_shipping = 'None'
            shipping.append(item_shipping)  

            try:
                item_stars = items.find('div', class_='b-starrating')
                item_rating = item_stars.find('span', class_='clipped').text.split(" ")[0]
            except Exception as e:
                item_rating = 'None'
            rating.append(item_rating)

            try:
                item_nreviews = items.find('span', class_='s-item__reviews-count').text.split(" ")[0]
            except Exception as e:
                item_nreviews = 'None' 
            review.append(item_nreviews)

            try:
                item_bids = items.find('span', class_='s-item__bids s-item__bidCount').text.split(" ")[0]
            except Exception as e:
                item_bids = 'None'
            bids.append(item_bids)
            
            category.append(j)
```

```python
dict = {'Category':category,'Title':title, 'Quality':sec,'Location':loc, 
        'Price-inr':price, 'Shipping-inr':shipping,'Stars':rating, 'Review-count':review,'Bids':bids}

prod_df = pd.DataFrame(dict)
prod_df.head(100)
```

```python
prod_df['Shipping-inr'] = prod_df['Shipping-inr'].replace(['not'],'0')
prod_df['Shipping-inr'] = prod_df['Shipping-inr'].replace(['International Shipping'],'0')
```

```python
prod_df= prod_df[prod_df.Title != 'None']
prod_df.drop(prod_df[(prod_df['Stars'] == 'None') & (prod_df['Review-count']=='None') & (prod_df['Bids']=='None')].index, inplace = True) 
```
